```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Male applicants, male admission rate, female applicants, female admission rate
# Each row contains data from one of the 6 largest depatments at Berkeley
# Data taken from https://en.wikipedia.org/wiki/Simpson%27s_paradox#UC_Berkeley_gender_bias
adms = [[825, 0.62, 108, 0.82], 
        [560, 0.63,  25, 0.68],
        [325, 0.37, 593, 0.34],
        [417, 0.33, 375, 0.35],
        [191, 0.28, 393, 0.24],
        [373, 0.06, 341, 0.07]]

adms = np.array(adms)
```

```python
def legend():
    plt.scatter([],[],alpha=0.7, c='C0', s=100, label='men')
    plt.scatter([],[],alpha=0.7, c='pink', s=100, label='women')
    plt.legend()
```

```python
plt.figure(figsize=(8,4))
grid = plt.GridSpec(1, 4, wspace=0.4, hspace=0.3)
plt.suptitle('Berkeley admission rates by gender (Fall, 1973)')

ax1 = plt.subplot(grid[0, :1])
plt.scatter([0], np.average(adms[:,1], weights=adms[:,0]), s=np.sum(adms[:,0]), alpha=0.7, c='C0')
plt.scatter([0], np.average(adms[:,3], weights=adms[:,2]), s=np.sum(adms[:,2]), alpha=0.7, c='pink')
plt.xticks([0], ['All'])
plt.xlabel('Departments')
plt.ylabel('Admission Rate')

plt.subplot(grid[0, 1:], sharey=ax1)
plt.scatter(np.arange(6), adms[:,1], s = adms[:,0], alpha=0.7, c='C0')
plt.scatter(np.arange(6), adms[:,3], s = adms[:,2], alpha=0.7, c='pink')
plt.xticks(np.arange(6), ['A','B','C','D','E','F'])
plt.xlabel('Departments')
legend()
```
